## Coding Session I - 30 minutes

From the Fitbit data, we want to visually understand: 

1. **How the average number of steps change by month. Use the appropriate visualization to show the median steps by month?**
1. **Is Fitbitter more active on weekend or workdays?**

##### All plots must be in the same jupyter notebook cell.

Hints:

- Use `Months_encoded` and `Week_or Weekend` columns
- Use `matplolib.pyplot` object oriented API
- Set your size figure to `12,4`
- Explore `plt.sca`
- Explore `plt.xticks`
- Save your figures

```python
# Driver 1 enter codes here
#para poder usar las funciones plt, en lugar de ax[n]. utilizamos plt.sca. 
#Razon? nombre.plot tiene mas opts de customizacion que ax[n].plot que solo admite (x,y)
#La funcion de sca es definir todo lo que esta por debajo como el grafico n indicado

group_months = fitbit.Steps.groupby(fitbit.Months_encoded).median()
group_week = fitbit.Steps.groupby(fitbit.Work_or_Weekend).median()


fig, ax = plt.subplots(1,2,figsize=(12,4))

plt.sca(ax[0])
group_months.plot(kind = 'line',lw=2, ls="--")
plt.ylabel('Median number of steps')
plt.xlabel('Months')
plt.title('Median Steps walked by month')
plt.xticks(list(range(13)),['','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])

plt.sca(ax[1])
group_week.plot(kind = 'bar',lw=2, ls="dashdot")
plt.ylabel('Median number of steps')
plt.xlabel('Weekday/Weekend')
plt.title('Median Steps walked by week/weekend')
plt.xticks(list(range(2)),['Weekend','Workdays'])
```

## Coding Session II - 30 minutes

**Write a loop to plot 3 scatter plots of the following features:**

1. **`Minutes Lightly Active`  vs `Steps`**
1. **`Minutes Very Active` vs `Steps`**
1. **`Minutes Sedentary` vs `Steps`**

```python
# Driver 2 enter codes here
df = fitbit[["Minutes Lightly Active", "Minutes Very Active", "Minutes Sedentary", "Steps"]]

fig, ax = plt.subplots(1, 3, figsize=(12,4))

for e in range(3):
    plt.sca(ax[e])
    plt.scatter(df.iloc[:,e],df['Steps'], alpha = 0.5)
    plt.xlabel(df.iloc[:,e].name)
    plt.ylabel('Steps')
```
